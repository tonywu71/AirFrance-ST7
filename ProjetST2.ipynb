### 1.3. Formalisme mathématique

#### 1.3.1. Ensembles principaux

Afin de formellement décrire l'ensemble des passagers de l'avion ainsi que les sièges de la cabine, introduisons les ensembles suivants :
<ul>
    <li> $\mathcal{P}$ l'ensemble des passagers ;
    <li> $\mathcal{S}_\emptyset$ l'ensemble des sièges vides ; 
    <li> $\mathcal{S} = [1,28] \times [1,7] \setminus \mathcal{S}_\emptyset$ l'ensemble des sièges réels.   
</ul>
A noter que si on voulait exprimer exhaustivement l'ensemble des sièges vides, on aurait donc : 
<br>
$$\mathcal{S}_\emptyset = ([ 1,28 ] \times \lbrace 4 \rbrace) \cup (\lbrace 1 \rbrace \times [ 5,8 ])$$

#### 1.3.2. Variables indicatrices

A partir de ces ensembles, on peut donc introduire les variables indicatrices $\delta_{xyp} \in \lbrace 0,1 \rbrace$ $(p \in \mathcal{P}, (x,y) \in [1,28] \times [1,7])$ valant $1$ si le passager $p$ est assis sur le siège de coordonnées $(x,y)$, et $0$ sinon.

Les variables $\delta_{xyp}$ nous serons très utiles pour exprimer des conditions liées à l'autorisation ou non que des passagers s'assoient sur un siège. Les conditions de base sont :
<ul>
    <li> Pas de passagers assis sur les sièges vides : $\sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}_\emptyset } \delta_{xyp} = 0$ ;
    <li> Au plus un passager par siège : $\forall (x,y) \in \mathcal{S}$, $\sum_{p \in \mathcal{P}} \delta_{xyp} \leq 1$ ;
    <li> Au plus un siège par passager : $\forall p \in \mathcal{P}$, $\sum_{(x,y) \in \mathcal{S}} \delta_{xyp} \leq 1$ ;
    <li> Tous les passagers ont un siège : $\sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}} \delta_{xyp} = \# \mathcal{P}$ ;   
</ul>

Ce modèle sera enrichi par la suite avec les différentes contraintes qui s'ajouteront à notre modèle. Par exemple, un détail important sur lequel nous nous attarderont subséquemment est le fait qu'aucun passager enfant ne doit être assis sur une ranngée située au niveau des issues de secours de l'avion. Sans s'épancher sur la définition formel des deux ensemble suivants, nous pouvons écrire :
<ul>
    <li> $\mathcal{E} \subset \mathcal{P}$ l'ensemble des passagers enfants ;
    <li> $\mathcal{S}_\mathcal{I} \subset \mathcal{S}$ l'ensemble des sièges situés au niveau des issues de secours.
</ul>
Dès lors, on aura : $\sum_{p \in \mathcal{E}} \sum_{(x,y) \in \mathcal{S}_\mathcal{I}} \delta_{xyp} = 0$, ce qui signifie bien qu'aucun enfant ne devra être assis près d'une issue de secours. Toutefois, cette condition ne sera envisagée que dans la seconde phase de notre étude donc nous y reviendrons ultérieurement.
<br> <br>
Cette partie avait juste pour but d'introduire formellement quelques ensembles d'intérêt ainsi que les variable $\delta_{xyp}$ qui nous serons très utiles par la suite, notamment en tant que variable de décision pour le problème d'optimisation.

## 2. Contraintes liées à la sécurité

### 2.1. Centrage de l'avion

Afin que l'avion puisse décoller sans risque et qu'il reste stable une fois en vol, le barycentre ce de dernier doit se situer dans une zone très particulière appelée "zone de centrage". Il s'agit d'une contrainte de sécurité forte devant être respectée obligatoirement. Cette zone se situe entre la $13^{ème}$ et la $17^{ème}$ rangée dans le sens de la longeur, et entre les deux colonnes bordant l'allée centrale dans le sens de la hauteur.

Dans le cadre de notre problème, on considère que le barycentre de l'avion à vide se situe effectivement dans cette zone. Il nous faut donc vérifier que la répartition des masses due au positionnement des passagers conserve le barycentre dans cette même zone de centrage.

#### 2.1.1. Formalisme mathématique

Considérons un ensemble de $N$ masses $(m_i)_{i \in [1,N]}$ réparties dans le plan. Pour chaque masse $m_i$, on désigne par $\vec{p_i}=(x_i,y_i)^T$ son vecteur position en partant de l'origine dudit plan. Dès lors, le vecteur position du barycentre $\vec{B}$ se calcule par :
$$\vec{B}=\frac{\sum_{i=1}^N m_i\vec{p_i}}{\sum_{i=1}^N m_i}$$
Dès lors, en projetant cette relation selon la direction $x$ et la direction $y$, on obtient deux équations découplées donnant la position barycentre :
<ul>
    <li> Positionnement selon $x$ : $B_x=\frac{\sum_{i=1}^N m_i x_i}{\sum_{i=1}^N m_i}$
    <li> Positionnement selon $y$ : $B_y=\frac{\sum_{i=1}^N m_i y_i}{\sum_{i=1}^N m_i}$
</ul>        

#### 2.1.2. Expression à l'aide des $\delta_{xyp}$

Compte tenu des définitions précédentes, on peut alors réécrire les conditions à l'aide des variables $\delta_{xyp}$ en considérant la direction $x$ définie ci-dessus comme la direction dans la longueur de l'avion, et $y$ celle dans la hauteur de ce dernier :
<ul>
    <li> Centrage selon la longueur : $13 \sum_{p \in \mathcal{P}} m_p \leq \sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}} m_p x \delta_{xyp} \leq 17 \sum_{p \in \mathcal{P}} m_p$
    <li> Centrage selon la largeur : $3 \sum_{p \in \mathcal{P}} m_p \leq \sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}} m_p y \delta_{xyp} \leq 5 \sum_{p \in \mathcal{P}} m_p$  
</ul>

## 3. Fonctions à maximiser

Dans le cadre de ce problème, on cherche à maximser le contentement des passagers d'un point de vue confort et pratique Pour cela, on cherche donc à faire en sorte que les passagers d'un même groupe soient assis à côté

### 3.1. Définition du voisinage

On considère que le voisinage d'un siège est constitué des huits sièges l'entourant dans le cas général. Lorsqu'il s'agit d'un siège à côté du couloir central, on considère que le siège de la même rangée de l'autre côté du couloir est aussi dans le voisinage.

Toutefois, il est évident que deux passagers du même groupe seront plus satisfaits s'ils sont côte-à-côte plutôt que séparés par un couloir, l'un devant l'autre ou en diagonale. Pour cela, on introduit différents coeffcients de contentement relatifs à la position relative de deux passagers, dont le plus grand est bien évidemment celui correspondant à la situation dans laquelle les passagers sont côte-à-côte :
<ul>
    <li> $c_{côté}$ lorsque les passagers sont côte-à-côte;
    <li> $c_{devant}$ lorsqu'un des deux passagers est assis devant l'autre;
    <li> $c_{diagonale}$ lorsque les deux passagers sont assis en diagonale l'un par rapport à l'autre;
    <li> $c_{couloir}$ lorsque les passagers sont séparés par le couloir central.
</ul>

### 3.2. Introduction de nouvelles variables de décision

On introduit donc, pour <i>mesurer</i> le score de contentement d'un passager vis-à-vis de son voisinnage, des variables $s_{xyp}$. Pour illustrer le fonctionnement de ces variables, donnons l'exemple d'un passager $p$ appartenant au groupe $G$ de taille inférieure à 3. Alors, on a : 
<br><br>
$$s_{xyp} = \sum_{q \in G \setminus \lbrace p \rbrace} c_{côté} ( \delta_{x,y+1,q} + \delta_{x,y-1,q}) 
+ c_{devant} (\delta_{x+1,y,q} + \delta_{x-1,y,q}) 
+ c_{diagonale} (\delta_{x+1,y+1,q} + \delta_{x-1,y+1,q} + \delta_{x+1,y-1,q} + \delta_{x-1,y-1,q})$$
<br>
Avec des mots : "on calcule le score par passager en sommant de manière pondérée la présence ou non de membres du même groupe dans le voisinnage du passager".
<br><br>
A noter qu'en l'abscence de passager sur le siège de coordonnées $(x,y)$, la valeur de $s_{xyp}$ est nécessairement nulle pour tout $p$. Formellement, $\delta_{xyp} = 0 \Rightarrow s_{xyp} = 0$.

### 3.3. Analyse des différents cas

#### 3.3.1. Siège situé côté couloir

Pour les sièges donnant sur le couloir, on rajoute un terme à la somme définissant $s_{xyp}$ et correspondant au poids de la présence d'un membre du groupe de l'autre côté du couloir. Ceci se produit donc pour les sièges tels que $y \in \lbrace 3,5 \rbrace$.

#### 3.3.2. Distinction sur la taille du groupe

Quand le groupe est constitué de trois personnes ou moins, idéalement, tout le groupe doit tenir sur trois sièges juxtaposés sur la même rangée. Dès lors, il serait contre-productif de favoriser le fait que les membres du groupe soient placés sur différentes rangées ou de part et d'autre du couloir. Ainsi, les coefficients $c_{devant}, c_{diagonale},$ et $c_{couloir}$ auront des valeurs différentes selon que le cardinal du groupe considéré est inférieur ou égal à trois, ou strictement supérieur à cette valeur (ces trois constantes étant vraisemblablement nulles ou très petites dans le cas des groupes à moins de trois membres).

#### 3.3.3. Introduction de sièges fictifs

On remarque que la formule définissant les $s_{xyp}$ peut être confrontée à différents cas pathologiques lorsque l'on considère un siège sur bord (par exemple, à côté du hublot). Pour pallier à cet écueil induit par ce que l'on pourrait appeler des <i>effets de bord</i>, on introduit donc des sièges fictifs sur la frontière des sièges existant (dont la valeur de $y$ vaut $0$ ou $8$ et celle de $x$ vaut $0$ ou $29$). On fera bien évidemment attention à ce que, pour tout siège vide et pour tout passager $p$, on ait bien : $\delta_{xyp} = 0$.

### 3.4. Utilisation des variables $s_{xyp}$

Les variables $s_{xyp}$ sont donc très utiles pour calculer le bonheur des passagers en fonction de leur voisinnage. En revanche, comme nous l'avons expliqué précédemment, si aucun passager n'est assis sur le siège $(x,y)$, la notion de voisinnage perd donc de son sens. Dès lors, il faut prendre en compte un nouveau type de variables de décisions que nous allons appeler les variables $\mu_{xyp}=min(M\delta_{xyp},s_{xyp})$, où $M$ est un majorant de l'ensemble des $s_{xyp}$; nous prendrons donc $M$ égal au nombre de passagers de l'avion.

De plus, mathématiquement, on a l'égalité :
$$\forall (a,b) \in \mathbb{R}^2 : min(a,b) = \frac{a+b-|a-b|}{2}$$

Ainsi en introduisant les variables de décision $a_{xyp}$ telles que $\forall (x,y,p) : a_{xyp} = |M \delta_{xyp} - s_{xyp}|$, on a :
<br><br>
$$\forall (x,y,p) : \mu_{xyp} = \frac{M \delta_{xyp} + s_{xyp} - a_{xyp}}{2}$$

### 4.2. Ensemble des passagers

Afin de simplifier l'utilisation des différentes données relatives aux passagers (âge, sexe, poids, classe, temps de correspondance, ...), nous avons fait le choix d'implémenter une classe objet Python pour les modéliser informatiquement. Cela permettra un appel à ces données plus aisé et commode. La classe ```Passager``` est définie ci-dessous. A noter que l'on définit tout d'abord deux classes énumérations afin de donner une liste exhaustive des différentes catégories de passagers ainsi que des différentes classes de voyageurs.

````python
class Passager:
    
    """
    Une classe représentant un passager et toutes les données qui lui sont associées. 
    """
    
    def __init__(self, idx, id_passager, categorie, classe, transit_time=0):
        
        """
        Constructeur pour la classe Passager.
        
        Args:
            idx (int): Indice du groupe dans le fichier Excel
            categorie (Categories): Catégorie du passager (Femme, Homme, Enfant, Fauteil Roulant)
            classe (Classes): Lettre donnant la classe du passager
            transit_time (string, optional): String au format %H:%M:%S donnat le
                temps de transit pour le passager. Defaults to 0.
        """
        
        self.idx = idx
        self.id_passager = id_passager
        self.categorie = categorie
        self.classe = classe
        self.transit_time = transit_time
        self.poids()
    
    def __str__(self):
        return f'Passager # {self.id_passager}, catégorie {self.categorie} du groupe #{self.idx}, classe {self.classe} et temps de correspondance de {self.transit_time}'
    
    def __repr__(self):
        return f'passager # {self.id_passager} du groupe #{self.idx}'          
    
    def poids(self):
        """
        Définit le poids du passager selon sa catégorie.
        """
        if self.categorie == "femmes":
            self.poids = 65
        elif self.categorie == "hommes":
            self.poids = 80
        else:
            self.poids = 35
````

````python
numPassager = 0
````

Dans cette même logique, on définit une classe ```Groupe``` pour décrire les groupes de passagers voyageant ensemble.

````python
class Groupe:
    
    """
    Une classe représentant un groupe de passagers ayant réservé leur places d'avion ensemble. 
    Un groupe correspond donc à une ligne dans le fichier Excel de départ.
    """

    def __init__(self, idx, nb_femmes, nb_hommes, nb_enfants, nb_WCHR, classe, transit_time=0):
        
        """
        Constructeur pour la classe Groupe.
        
        Args:
            idx (int): Indice du groupe dans le fichier Excel
            nb_femmes (int): Nombre de femmes dans le groupe
            nb_hommes (int): Nombre d'hommes dans le groupe
            nb_enfants (int): Nombre d'enfants dans le groupe
            nb_WCHR (int): Nombre de personnes handicapés dans le groupe
            classe (Classes): Lettre donnant la classe du groupe
            transit_time (string, optional): String au format %H:%M:%S donnat le
                temps de transit pour le groupe. Defaults to 0.
        """
        
        self.idx = idx
        self.composition = {
            'femmes': nb_femmes,
            'hommes': nb_hommes,
            'enfants': nb_enfants,
            'WHCR': nb_WCHR
        }
        self.classe = classe
        self.transit_time = transit_time

        self.list_passagers = []
        global numPassager

        #for categorie, nombre in self.composition.items():
        for categorie in self.composition.keys():
            for nbParCategorie in range(self.composition[categorie]):
                self.list_passagers.append(Passager(
                    self.idx,
                    numPassager,
                    categorie,
                    self.classe,
                    self.transit_time))
                numPassager += 1   

        return

    def __str__(self):
        return f'Groupe #{self.idx} avec {self.get_nombre_passagers()} passager(s), classe {self.classe} et temps de correspondance de {self.transit_time}'

    def __repr__(self):
        return f'groupe #{self.idx}'
    
    def iter_passagers(self):
        """Générateur pour la liste des passagers dans le groupe.
        Yields:
            Passager: instance de la classe Passager
        """
        for passager in self.list_passagers:
            yield passager

    def est_seul(self):
        """Renvoie True si le groupe contient un seul
        passager et False sinon.
        """
        return len(self.list_passagers) == 1
    
    def get_nombre_passagers(self):
        """Renvoie le nombre de passagers dans le groupe considéré.
        """
        return len(self.list_passagers)

    def comprend_enfants(self):
        """Renvoie True si le groupe contient au moins un enfant
        en son sein et False sinon.
        """
        return any([passager.categorie == 'enfants' for passager in self.list_passagers])      
````

### Export du résultat en CSV

````python
def exporter_resultats(listeGroupes, placement):    
    list_id_groupe = []
    list_id_passager = []
    list_categorie = []
    list_classe = []
    list_transit_time = []
    list_poids = []
    list_x = []
    list_y = []
    list_siege = []
    list_classes = []
    
    for groupe in listeGroupes.keys():        
        for passager in listeGroupes[groupe].list_passagers:
            x, y = placement[passager.id_passager]
            siege = coordToSiege(x, y)
            
            list_id_groupe.append(listeGroupes[groupe].idx)
            list_id_passager.append(passager.id_passager)
            list_categorie.append(passager.categorie)
            list_classe.append(passager.classe)
            list_transit_time.append(passager.transit_time)
            list_poids.append(passager.poids)
            list_x.append(x)
            list_y.append(y)
            list_siege.append(siege)
            list_classes.append(get_classe(x, y))
    
    df = pd.DataFrame({
        'ID Groupe': list_id_groupe,
        'ID Passager': list_id_passager,
        'Catégorie': list_categorie,
        'Classe': list_classe,
        'Transit Time': list_transit_time,
        'Poids': list_poids,
        'x': list_x,
        'y': list_y,
        'Siège': list_siege,
        'Classe': list_classes
    })
    
    return df
````
